# EDA

# Gender

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for var in ax.containers:
    ax.bar_label(var)
```

```python
df['Amount'].sum()
```

```python
x=df.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(x='Gender',y='Amount',data=x)
```

from this graph we can clearly undertand that most of the customers and and the purchase more than male

# Age Group

```python
df.columns
```

```python
sns.countplot(x='Age Group',data=df)
```

```python
sns.countplot(x='Age Group',hue='Gender',data=df)
```

```python
a=sns.countplot(x='Age Group',hue='Gender',data=df)
for var in a.containers:
    a.bar_label(var)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum()
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from the above graph we conclde that the 26-35 age group buy the most and from that females are more

# state

```python
df.columns
```

```python
state_sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(7)
```

```python
sns.barplot(data=state_sales,x='State',y='Orders')
```

here we can notice that our map is collapsed with each other

```python
state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(15)
sns.set(rc={'figure.figsize':(25,11)})

sns.barplot(data=state_orders,x='State',y='Orders')
```

```python
state_sale=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(25,11)})
sns.barplot(data=state_sale,x='State',y='Amount')
```

from the above graph we can conclde that most of the orders are from up,maharastra,karnataka

# Marital Status

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(3,6)})
a=sns.countplot(x='Marital_Status',data=df)
for var in a.containers:
    a.bar_label(var)
```

```python
marital_sale = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
a=sns.barplot(data=marital_sale, x='Marital_Status', y='Amount', hue='Gender')
for var in a.containers:
    a.bar_label(var)
```

from the above graph we can conclude that most of the buyer are married women and they have more purchasing power 

# occupation

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(28,9)})
a=sns.countplot(data=df,x='Occupation')
for var in a.containers:
    a.bar_label(var)
```

```python
occupation_sale=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
a=sns.barplot(data=occupation_sale,x='Occupation',y='Amount')
for var in a.containers:
    a.bar_label(var)
```

from the above graph we can conclude that most of the buyers are working at it sector and healthcare

# Product category 

```python
df.columns
```

```python
sns.countplot(x='Product_Category',data=df)
```

```python
sns.set(rc={'figure.figsize':(35,10)})
product_sale = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
a=sns.barplot(x='Product_Category', y='Amount', data=product_sale)
for var in a.containers:
    a.bar_label(var)
```

from this graph we say that most of the amount received from food,clothing and electronics

```python
sns.set(rc={'figure.figsize':(35,10)})
product_sale = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
a=sns.barplot(x='Product_Category', y='Orders', data=product_sale)
for var in a.containers:
    a.bar_label(var)
```

from the graph we conclude that most of the orders are from clothing,food and electronics

# CONCUSION

The analysis of Diwali sales data reveals that married women aged 26-35 are the primary purchasers, with Uttar Pradesh, Maharashtra, and Karnataka being the top contributing states. Clothing, food, and electronics emerge as the most popular product categories, suggesting targeted marketing efforts and product diversification opportunities.
